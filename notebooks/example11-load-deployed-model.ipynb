```python
import pandas as pd
from joblib import load  # fungsi untuk memuat (load) model atau objek .joblib yang sebelumnya disimpan
```

```python
model = load('../models/model_knn.joblib')
```

```python
sample = [5.9, 3.0, 5.1, 1.8]
formatted_sample = pd.DataFrame([sample], columns=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
pred = model.predict(formatted_sample)
print(pred)
```
