```python
import pandas as pd
from sklearn.model_selection import cross_val_score, KFold, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris_data = pd.read_csv("../data/Iris.csv")
```

```python
X = iris_data[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y = iris_data["Species"]
```

```python
model_knn = KNeighborsClassifier(n_neighbors=7)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model_knn, X, y, cv=kf, scoring="accuracy")
print("Cross-validation scores (accuracy tiap fold menggunakan KFold):")
print(cv_scores)
```

```python
print("Rata-rata akurasi: ", cv_scores.mean())
print("Standar Deviasi: ", cv_scores.std())
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(model_knn, X, y, cv=skf, scoring="accuracy")
print("Cross-validation scores (accuracy tiap fold menggunakan StratifiedKFold):")
print(cv_scores)
```

```python
print("Rata-rata akurasi: ", cv_scores.mean())
print("Standar Deviasi: ", cv_scores.std())
```
