```python
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import MinMaxScaler
```

```python
wine_data = load_wine()
```

```python
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
wine_df["target"] = wine_data.target
```

```python
X = wine_df[wine_data.feature_names].copy()
y = wine_df["target"].copy()
```

```python
# Standardisasi fitur menggunakan MinMaxScaler.
# MinMaxScaler akan mengubah setiap fitur ke rentang [0, 1]
# dengan rumus :
#     X_norm = (x - min) / (max - min)
#
# 1) scaler.fit(X)
#      -> menghitung nilai minimum dan maksimum dari setiap kolom fitur
# 2) scaler.transform(X)
#      -> menormalkan data sesuai rumus MinMaxScaler

                # ┌────────────────────────────┐
                # │      Data Fitur X          │
                # │  (DataFrame: alkohol, dll) │
                # └──────────────┬─────────────┘
                #                │
                #                ▼
                #  ┌─────────────────────────┐
                #  │    scaler = MinMaxScaler│
                #  └─────────────┬───────────┘
                #                │
                #                ▼
                #  ┌─────────────────────────┐
                #  │       scaler.fit(X)     │
                #  │  - Hitung min tiap kolom│
                #  │  - Hitung max tiap kolom│
                #  └─────────────┬───────────┘
                #                │
                #                ▼
                #  ┌─────────────────────────┐
                #  │   scaler.transform(X)   │
                #  │  - Setiap nilai dihitung│
                #  │    (x - min) / (max-min)│
                #  └─────────────┬───────────┘
                #                │
                #                ▼
                #  ┌──────────────────────────┐
                #  │     X_scaled (array)     │
                #  │  Data sudah dinormalisasi│
                #  │    range semua fitur     │
                #  │       = [0, 1]           │
                #  └──────────────────────────┘


# Scaling dengan MinMaxScaler
scaler = MinMaxScaler()

# 1. Fit = hitung nilai minimum dan maksimum tiap fitur
scaler.fit(X)

# 2. Transform = ubah data X menggunakan min & max hasil fit
X_scaled_array = scaler.transform(X)

# 3. Ubah hasil scaling menjadi DataFrame agar mudah dibaca
X_scaled = pd.DataFrame(X_scaled_array, columns=wine_data.feature_names)
```
